```python
#Importing Libraries
import sys
import pandas as pd 
import numpy as np
import sklearn
import matplotlib
```

```python
#Importing Iris dataset from sklearn
from sklearn import datasets
iris = datasets.load_iris()
features = iris.data
target = iris.target

#Generating pandas dataframe
df = pd.DataFrame(features)
df.columns = iris.feature_names
```

```python
#Print out target names
print(target)
iris.target_names
```

```python
#Print dataset information
print(df.shape)
print(df.head(20))
# 4 Features 150 Instances
```

```python
#Print dataset descriptions and class distributions
print(df.describe())
```

```python
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt

#Display scatter plot matrix
scatter_matrix(df)
plt.show()

#data good for clustering due to seperation of data
#good for PCA analysis due to correlation between data
```

```python
#Elbow method to determine optimal number of clusters
from sklearn.cluster import KMeans

#empty x and y data lists
X = []
Y = []

#loop through different number of clusters
for i in range(1, 31):
    #initiate and fit the model
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(df)
    
    #append number of clusters to x data list
    X.append(i)
    
    #append average within cluster sum of squares to y data list
    awcss = kmeans.inertia_ / df.shape[0]
    Y.append(awcss)
```

```python
import matplotlib.pyplot as plt

#plot the x and y axis
plt.plot(X,Y, 'bo-')
plt.xlim((1, 30))
plt.xlabel('Number of Clusters')
plt.ylabel('Average within-cluster sum of squares')
plt.title('K-Means Clustering Elbow Method')

#display the plot
plt.show()

#3 clusters is optimal as seen on the graph
```

```python
#PCA is best use which features that have correlations
from sklearn.decomposition import PCA
from sklearn import preprocessing
```

```python
#performing PCA, reducing to 2 components
pca = PCA(n_components = 2)
pc = pca.fit_transform(df)

#print new dimensions
print (pc.shape)
print (pc[:10])
```

```python
#refit kmeans model to  the principal compoenent with the appropriate number of clusters
kmeans = KMeans(n_clusters = 3)
kmeans.fit(pc)
```

```python
#Visualize high dimensional clusters using principle components

#set size for the mesh
h = 0.02

#genrating mesh grid
x_min, x_max = pc[:, 0].min() - 1, pc[:, 0].max() + 1
y_min, y_max = pc[:, 1].min() - 1, pc[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

#Label each point in mesh using last trained model
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])

#generate color plot from results
Z = Z.reshape(xx.shape)
plt.figure(figsize = (12, 12))
plt.clf()
plt.imshow(Z, interpolation = 'nearest', extent = (xx.min(), xx.max(), yy.min(), xx.max()), cmap = plt.cm.tab20c, aspect = 'auto', origin = 'lower')

#plot the principle components on the color plot
for i, point in enumerate(pc):
    if target[i] == 0:
        plt.plot(point[0], point[1], 'g.', markersize = 10)
    if target[i] == 1:
        plt.plot(point[0], point[1], 'r.', markersize = 10)
    if target[i] == 2:
        plt.plot(point[0], point[1], 'b.', markersize = 10)
#plot the cluster centorids
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], marker = 'x', s=250, linewidth = 4, color = 'w', zorder = 10)

#set plot title and axis limits
plt.title('K-Means Clustering on PCA-Reduced Iris Data Set')
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.xticks(())
plt.yticks(())

#display
plt.show()

#possible to change to 4 clusters but the result is not that good
```

```python
#3 Metrics to classify the success of our classification
#Homogeneity - measures whether or not all of its clusters contain only data points which are members of a single class
#Completeness - measures whether or not all members of a given class are elements of the same cluster
#V-Measure - the harmonic mean between homogeneity and completeness
```

```python
from sklearn import metrics

# K means clustering on non reduced data
kmeans1 = KMeans(n_clusters = 3)
kmeans1.fit(features)

# K means clustering on PCA reduced data
kmeans2 = KMeans(n_clusters = 3)
kmeans2.fit(pc)

#print metrics for non reduced data
print('Non-reduced data')
print('Homogeneity: {}'.format(metrics.homogeneity_score(target, kmeans1.labels_)))
print('Completeness: {}'.format(metrics.completeness_score(target, kmeans1.labels_)))
print('V-measure: {}'.format(metrics.v_measure_score(target, kmeans1.labels_)))

#print metrics for PCA reduced data
print('PCA reduced data')
print('Homogeneity: {}'.format(metrics.homogeneity_score(target, kmeans2.labels_)))
print('Completeness: {}'.format(metrics.completeness_score(target, kmeans2.labels_)))
print('V-measure: {}'.format(metrics.v_measure_score(target, kmeans2.labels_)))
```

```python
#Slight difference in results for PCA reduced data which was expected
#Compare results further, print out the actual labels
print (kmeans1.labels_)
print (kmeans2.labels_)
print (target)

#Finding
#Some misclassification in cluster 2 and 3 which was supported by the graph above
```
